```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("insurance 2 (2).csv")
df.head()
```

```python
## Univariate Analysis
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

 Observation:
The age distribution is slightly right-skewed, with most individuals between 20 and 40 years old.
This indicates that the dataset mainly represents young and middle-aged adults.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["bmi"], kde=True)
plt.title("BMI Distribution")
plt.show()
```

 Observation:
BMI values are normally distributed, with most individuals having a BMI between 25 and 35.
This suggests that a large portion of the dataset is in the overweight range.

```python
sns.countplot(x="smoker", data=df)
plt.title("Smokers Count")
plt.show()
```

 Observation:
The number of non-smokers is significantly higher than smokers in the dataset.
This indicates an imbalance between the two categories.

```python
## Bivariate Analysis
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="smoker", y="charges", data=df)
plt.title("Charges vs Smoker")
plt.show()
```

 Observation:
Smokers have significantly higher medical charges compared to non-smokers.
There are also more extreme values among smokers, showing that smoking greatly increases insurance costs.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="age", y="charges", data=df)
plt.title("Age vs Charges")
plt.show()
```

 Observation:
There is a positive relationship between age and medical charges.
As age increases, the insurance cost tends to increase.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="bmi", y="charges", data=df)
plt.title("BMI vs Charges")
plt.show()
```

 Observation:
There is a moderate positive relationship between BMI and medical charges.
Individuals with higher BMI tend to have higher insurance costs.

```python
## Correlation
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

 Observation:
Age and charges show the strongest positive correlation.
BMI has a moderate correlation with charges, while the number of children has a weak relationship with medical costs.

## Conclusion

- The dataset contains 1338 records and 7 features.
- There are no missing values in the dataset.
- Most individuals are non-smokers.
- Medical charges are highly affected by smoking habits.
- Age has a positive relationship with medical charges.
- BMI shows a moderate effect on insurance costs.
- Smokers pay significantly higher medical costs compared to non-smokers.
